```python
import pandas as pd

```

```python
df = pd.read_csv("UK_crime_datasets.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()

df = df.dropna(subset=["Crime type"])
```

```python
df["Month"] = pd.to_datetime(df["Month"], format="%Y-%m")
```

```python
len(df)
```

```python
df["Crime type"].value_counts()
```

```python
df["Falls within"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df["Crime type"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Crime Types in the UK")
plt.show()
```

```python
df.groupby("Month").size().plot()
plt.title("Crime Trend Over Time")
plt.show()
```
